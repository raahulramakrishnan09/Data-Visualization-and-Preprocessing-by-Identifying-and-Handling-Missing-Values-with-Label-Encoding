```python
#finding_outliers_using_for_loop
plt.figure(figsize=(30,25))
plotnumber=1

for col in df.columns:
  if plotnumber<=24:
    ax=plt.subplot(5,5,plotnumber)
    sns.boxplot(df[col])
    plt.xlabel(col,fontsize=25)

  plotnumber +=1
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20,15))
plotnumber=1

for col in df.columns:
  if plotnumber<=24:
    ax=plt.subplot(5,5,plotnumber)
    sns.histplot(df[col])
    plt.xlabel(col,fontsize=15)

  plotnumber +=1
plt.tight_layout()
plt.show()
```
